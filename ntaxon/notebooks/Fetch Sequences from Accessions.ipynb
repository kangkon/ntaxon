```python
import pandas as pd
from Bio import Entrez, SeqIO
```

```python
Entrez.email='kangkonsaikia@gmail.com'

accessions = pd.read_csv('./data/shabiha_bacteria_accessions.csv').set_index("accession")

accessions.head()
```

```python
def fetch_sequence(row):
    # get Index - row.name 
    handle = Entrez.efetch(db='nucleotide', id=row.name, rettype='fasta')
    fasta = SeqIO.read(handle, 'fasta')
    return str(fasta.seq)
    
accessions['sequence'] = accessions.apply(lambda r : fetch_sequence(r), axis=1)
accessions.head()
```

```python
accessions.to_csv('./outputs/shabiha_bacteria_sequences.csv')
```
